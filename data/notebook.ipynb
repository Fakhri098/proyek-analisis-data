## Pengaruh Faktor Lingkungan terhadap Penyewaan Sepeda?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Asumsi all_df memiliki kolom 'dteday' untuk tanggal, 'cnt' untuk jumlah penyewaan,
# 'temp' untuk suhu, 'hum' untuk kelembaban, dan 'windspeed' untuk kecepatan angin

# Menambahkan kolom bulan dalam all_df
all_df['month'] = pd.to_datetime(all_df['dteday']).dt.month

# Melakukan RFM sederhana dengan pengelompokan bulanan
rfm_df = all_df.groupby('month').agg({
    'cnt': 'sum',
    'temp': 'mean',
    'hum': 'mean',
    'windspeed': 'mean'
}).reset_index()

# Menentukan ukuran figure dan gaya untuk visualisasi yang konsisten
sns.set(style="whitegrid")
plt.figure(figsize=(10, 15))

# Scatter plot antara suhu dan jumlah penyewaan dengan garis regresi untuk mempermudah pemahaman tren
plt.subplot(3, 1, 1)
sns.regplot(data=all_df, x='temp', y='cnt', scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Hubungan antara Suhu dan Jumlah Penyewaan (Harian)')
plt.xlabel('Suhu (C)')
plt.ylabel('Jumlah Penyewaan')
plt.tight_layout()

# Scatter plot antara kelembaban dan jumlah penyewaan dengan garis regresi
plt.subplot(3, 1, 2)
sns.regplot(data=all_df, x='hum', y='cnt', scatter_kws={'alpha':0.3}, line_kws={'color':'blue'})
plt.title('Hubungan antara Kelembaban dan Jumlah Penyewaan (Harian)')
plt.xlabel('Kelembaban (%)')
plt.ylabel('Jumlah Penyewaan')
plt.tight_layout()

# Scatter plot antara kecepatan angin dan jumlah penyewaan dengan garis regresi
plt.subplot(3, 1, 3)
sns.regplot(data=all_df, x='windspeed', y='cnt', scatter_kws={'alpha':0.3}, line_kws={'color':'green'})
plt.title('Hubungan antara Kecepatan Angin dan Jumlah Penyewaan (Harian)')
plt.xlabel('Kecepatan Angin')
plt.ylabel('Jumlah Penyewaan')
plt.tight_layout()
plt.show()

# Visualisasi RFM sederhana: Bulanan Jumlah Penyewaan dan Kondisi Cuaca Rata-rata
plt.figure(figsize=(12, 6))
sns.barplot(data=rfm_df, x='month', y='cnt', color='#4a90e2', label='Total Penyewaan')
plt.plot(rfm_df['month'] - 1, rfm_df['temp'] * 100, color='red', marker='o', linestyle='--', label='Rata-rata Suhu x 100')
plt.plot(rfm_df['month'] - 1, rfm_df['hum'] * 100, color='blue', marker='o', linestyle='--', label='Rata-rata Kelembaban x 100')
plt.plot(rfm_df['month'] - 1, rfm_df['windspeed'] * 100, color='green', marker='o', linestyle='--', label='Rata-rata Kecepatan Angin x 100')
plt.title("Analisis RFM Sederhana: Bulanan Penyewaan dan Kondisi Cuaca Rata-rata")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penyewaan")
plt.legend()
plt.tight_layout()
plt.show()
```
